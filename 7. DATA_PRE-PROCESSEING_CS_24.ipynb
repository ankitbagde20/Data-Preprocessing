## 2.	Data Manipulation:

### a.	Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.

[Hint: Use sum method with appropriate axis]


```python
prisoners = prisoner.copy()
prisoners['Total benefitted'] = prisoners.iloc[:, 2:].sum(axis = 1)
prisoners
```

### b. Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states.

```python
prisoners_total = prisoners.append(prisoners.iloc[:, 2:].sum(), ignore_index = True)
prisoners_total
```

## 3.	Plotting:

### a.	Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

[Hint: Use bar method of pyplot]



```python
xlabels = prisoners['STATE/UT'].values

plot.figure(figsize=(20, 10))
plot.xticks(np.arange(xlabels.shape[0]),xlabels,rotation='vertical', fontsize=18)
plot.xticks

plot.bar(np.arange(prisoners.values.shape[0]),prisoners['Total benefitted'], edgecolor = 'r')
plot.show()
```

### b.	Make a pie chart that depicts the ratio among different modes of benefits.

```python
labels = prisoners.columns[2:-1]
sizes = prisoners_total.iloc[-1][2:-1].values
fig1, ax1 = plot.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plot.show()
```
